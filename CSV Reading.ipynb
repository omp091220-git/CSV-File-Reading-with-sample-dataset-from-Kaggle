## **Method-1**

```python
#single column
csv_file_reading_3 = pd.read_csv('https://raw.githubusercontent.com/omp091220-git/CSV-File-Reading-with-sample-dataset-from-Kaggle/refs/heads/main/Datasets/host_city_context.csv', usecols=["city"])
csv_file_reading_3
```

## **Method-2**

There are two sub-methods in this way:
1) In first sub-method, a seperate copy is created and in that, your desired column is stored. This means there's no change in the original variable.
2) In second sub-method, you make the changes in your main reading variable, which means that just like when we do 'sum=sum+5', same things happen in this case.

```python
#Sub-Method:1
csv_file_reading_5 = pd.read_csv('https://raw.githubusercontent.com/omp091220-git/CSV-File-Reading-with-sample-dataset-from-Kaggle/refs/heads/main/Datasets/host_city_context.csv')
csv_file_reading_6 = csv_file_reading_5['city']
#csv_file_reading_5
csv_file_reading_6
```

```python
#Sub-Method:2
csv_file_reading_5 = pd.read_csv('https://raw.githubusercontent.com/omp091220-git/CSV-File-Reading-with-sample-dataset-from-Kaggle/refs/heads/main/Datasets/host_city_context.csv')
csv_file_reading_5 = csv_file_reading_5['city']
csv_file_reading_5
```
